Mask both input and output

```python
def redact_pii(data, **kwargs):
    if isinstance(data, str):
        for key, pattern in PII_PATTERNS.items():
            data = re.sub(pattern, f"[{key}_REDACTED]", data)
        return data
    elif isinstance(data, dict):
        return {k: redact_pii(v) for k, v in data.items()}
    elif isinstance(data, (list, tuple)):
        return type(data)(redact_pii(x) for x in data)
    return data

Langfuse(mask=redact_pii)
```
